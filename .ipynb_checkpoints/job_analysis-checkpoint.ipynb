```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset("lukebarousse/data_jobs")

df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df
```

```python
df_filled = df.copy()

df_filled.loc[:, 'salary_rate':'salary_hour_avg'].dropna(subset = 'salary_rate')
median_salary_year = df_filled['salary_year_avg'].median()
median_salary_hour = df_filled['salary_hour_avg'].median()

df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_salary_year)
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_salary_hour)
```

```python
df_filled.loc[:10, 'salary_year_avg':'salary_hour_avg']
```

```python
df_unique = df_filled

df_unique = df_unique.drop_duplicates()
```

```python
df_unique = df_unique.drop_duplicates(subset = ['job_title', 'company_name'])
```

```python
#sample items to see what the data is representative of
df_unique.sample(10)
```

```python
df_jc_salary = df.pivot_table(values = "salary_year_avg", index = 'job_country', columns = 'job_title_short', aggfunc = "median")
```

```python
countries = df['job_country'].value_counts().head(10).index

df_jc_salary = df_jc_salary.loc[countries]
titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']
df_jc_salary = df_jc_salary[titles]
```

```python
df_jc_salary.plot(kind = "bar")
plt.xlabel("countries")
plt.ylabel("median salary in USD")
plt.title("Median Salary by country for top 3 jobs")
plt.legend()
```

```python
df_ind = df[df['job_country'] == 'India'].copy()
```

```python
df_ind['job_posted_month'] = df_ind['job_posted_date'].dt.strftime('%B')
```

```python
df_ind_pivot = df_ind.pivot_table(index = 'job_posted_month', columns = 'job_title_short', aggfunc = 'size')

df_ind_pivot.reset_index(inplace = True)
df_ind_pivot['month number'] = pd.to_datetime(df_ind_pivot['job_posted_month'], format = '%B').dt.month

df_ind_pivot.sort_values('month number', inplace = True)
df_ind_pivot.set_index('job_posted_month', inplace = True)
df_ind_pivot.drop(columns = 'month number', inplace = True)

df_ind_pivot
```

```python
# the graph is very cluttered so filter it further to only do top 3 jobs
df_ind_pivot.plot(kind = 'line')
```

```python
top_3 = df_ind['job_title_short'].value_counts().head(3).index.tolist()
```

```python
df_ind_pivot[top_3].plot(kind = 'line')
plt.xlabel("Year = 2025")
plt.ylabel("Job postings")
plt.title("Job postings by month for top 3 most posted jobs")
```

```python
import ast

df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
skills_count = df_exploded.groupby(['job_title_short', 'job_skills']).size()
skill_count = skills_count.reset_index(name = 'skill_count').sort_values(by = 'skill_count', ascending = False)
```

```python
df_exploded = df_DA.explode('job_skills')
skill_stats = df_exploded.groupby('job_skills').agg(
    skill_count = ('job_skills', 'count'),
    median_salary = ('salary_year_avg', 'median')
)

skill_stats = skill_stats.sort_values(by='skill_count', ascending = False).head(10)
skill_stats.plot(kind = 'scatter', x = 'skill_count', y = 'median_salary')
plt.xlabel('Count of job postings')
plt.ylabel('Median Yearly Salary')
plt.title('Salary vs Job postings for top 10 skills')

for i, txt in enumerate(skill_stats.index):
    plt.text(skill_stats['skill_count'].iloc[i],skill_stats['median_salary'].iloc[i], txt)
```

```python
job_title = ['Data Analyst', "Data Engineer", 'Data Scientist']
df_india = df[(df['job_title_short'].isin(job_title)) & (df['job_country'] == "India")].copy()

df_india = df_india.dropna(subset = ['salary_year_avg'])

job_list = [df_india[df_india['job_title_short'] == jt]['salary_year_avg']for jt in job_title]

plt.boxplot(job_list, labels = job_title, vert = False)
plt.xlabel('Salary dist in India')
plt.title('Yearly Salary')
```

```python
df_ind_DA = df_ind_DA.explode('job_skills')

df_ind_DA_group = df_ind_DA.groupby('job_skills')['salary_year_avg'].agg(['count', 'median'])

df_DA_top_pay = df_ind_DA_group.sort_values(by='median', ascending = False).head(10)
df_DA_skills = df_ind_DA_group.sort_values(by='count', ascending = False).head(10).sort_values(by='median', ascending = False)
```

```python
fig, ax = plt.subplots(2,1)

df_DA_top_pay.plot(kind = 'barh', y='median', ax = ax[0], legend = False)
ax[0].invert_yaxis()
df_DA_skills[::-1].plot(kind = 'barh', y='median', ax = ax[1], legend = False)
ax[1].set_xlim(ax[0].get_xlim())

ax[0].set_title("Top Paying skills in India for Data Analysts")
ax[1].set_title("Top in demand skills in India for Data Analysts")
ax[0].set_xlabel("")
ax[1].set_xlabel("Salary in USD")
ax[0].set_ylabel("")
ax[1].set_ylabel("")

fig.tight_layout()
```
